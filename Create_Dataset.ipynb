```python
import pandas as pd
import re
```

```python
def converttolower(texts):
  corpus = []
  for line in texts:
    line = line.lower()
    corpus.append(line)
  return corpus

def rempunct(texts):
  corpus = []
  for line in texts:
    opt = re.sub(r'[^\w\s]', '', line)
    corpus.append(opt)
    
  return corpus

def addspaces(tokens):
  words = []
  for token in tokens:
    temp = " " + token + " "
    words.append(temp)
  return words

def make_processable(texts):
  corpus = rempunct(texts)
  corpus = addspaces(corpus)
  corpus = converttolower(corpus)
  return corpus

def define_datapoint(dataframe, features):
  input_data = []

  for datapoint in dataframe:
    temp = []
    for feature in features:
      count = datapoint.count(feature)
      if count > 0:
        temp.append(count)
      else:
        temp.append(0)
    input_data.append(temp)

  return input_data

def compile_datapoints(*argv):
  final_features = []
  for arg in argv:
    for token in arg:
      if not token in final_features:
        final_features.append(token)
  return final_features
```
